```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import statistics as stat

# Dataframe tagging structure
def tag_df(df, language):
    df = df.copy()
    df["Language"] = language
    return df

# Constants
file_name_CS_Deb_GPIO = "dataGpio_C#_Debug.csv"
file_name_CS_Deb_Serial = "dataSerial_C#_Debug.csv"
file_name_CS_Rel_GPIO = "dataGpio_C#_Release.csv"
file_name_CS_Rel_Serial = "dataSerial_C#_Release.csv"
file_name_C_GPIO = "dataGpio_C.csv"
file_name_C_Serial = "dataSerial_C.csv"
file_name_Bash_GPIO = "dataGpio_Bash.csv"
file_name_Bash_Serial = "dataSerial_Bash.csv"
file_name_Python_GPIO = "dataGpio_Python.csv"
file_name_Python_Serial = "dataSerial_Python.csv"
file_name_Rust_GPIO = "dataGpio_Rust.csv"
file_name_Rust_Serial = "dataSerial_Rust.csv"
file_name_CS_Deb_Remeas_GPIO = "dataGpio_C#_Debug_Remeasured.csv"
file_name_CS_Deb_Remeas_Serial = "dataSerial_C#_Debug_Remeasured.csv"

# Load data from CSV file

# C# Debug GPIO
df_CS_Deb_GPIO = pd.read_csv(file_name_CS_Deb_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Deb_GPIO['Datetime'] = pd.to_datetime(df_CS_Deb_GPIO['Datetime'])
df_CS_Deb_GPIO["Language"] = "C# Debug GPIO"
# C# Debug Serial
df_CS_Deb_Serial = pd.read_csv(file_name_CS_Deb_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Deb_Serial['Datetime'] = pd.to_datetime(df_CS_Deb_Serial['Datetime'])
df_CS_Deb_Serial['Result_ms'] = df_CS_Deb_Serial['Result_ms'] - 0.9375
df_CS_Deb_Serial["Language"] = "C# Debug Serial"
# C# Release GPIO
df_CS_Rel_GPIO = pd.read_csv(file_name_CS_Rel_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Rel_GPIO['Datetime'] = pd.to_datetime(df_CS_Rel_GPIO['Datetime'])
df_CS_Rel_GPIO["Language"] = "C# Release GPIO"
# C# Release Serial
df_CS_Rel_Serial = pd.read_csv(file_name_CS_Rel_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Rel_Serial['Datetime'] = pd.to_datetime(df_CS_Rel_Serial['Datetime'])
df_CS_Rel_Serial['Result_ms'] = df_CS_Rel_Serial['Result_ms'] - 0.9375
df_CS_Rel_Serial["Language"] = "C# Release Serial"
# C GPIO
df_C_GPIO = pd.read_csv(file_name_C_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_C_GPIO['Datetime'] = pd.to_datetime(df_C_GPIO['Datetime'])
df_C_GPIO["Language"] = "C GPIO"
# C Serial
df_C_Serial = pd.read_csv(file_name_C_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_C_Serial['Datetime'] = pd.to_datetime(df_C_Serial['Datetime'])
df_C_Serial['Result_ms'] = df_C_Serial['Result_ms'] - 0.9375
df_C_Serial["Language"] = "C Serial"
# Bash GPIO
df_Bash_GPIO = pd.read_csv(file_name_Bash_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Bash_GPIO['Datetime'] = pd.to_datetime(df_Bash_GPIO['Datetime'])
df_Bash_GPIO["Language"] = "Bash GPIO"
# Bash Serial
df_Bash_Serial = pd.read_csv(file_name_Bash_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Bash_Serial['Datetime'] = pd.to_datetime(df_Bash_Serial['Datetime'])
df_Bash_Serial['Result_ms'] = df_Bash_Serial['Result_ms'] - 0.9375
df_Bash_Serial["Language"] = "Bash Serial"
# Python GPIO
df_Python_GPIO = pd.read_csv(file_name_Python_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Python_GPIO['Datetime'] = pd.to_datetime(df_Python_GPIO['Datetime'])
df_Python_GPIO["Language"] = "Python GPIO"
# Python Serial
df_Python_Serial = pd.read_csv(file_name_Python_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Python_Serial['Datetime'] = pd.to_datetime(df_Python_Serial['Datetime'])
df_Python_Serial['Result_ms'] = df_Python_Serial['Result_ms'] - 0.9375
df_Python_Serial["Language"] = "Python Serial"
# Rust GPIO
df_Rust_GPIO = pd.read_csv(file_name_Rust_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Rust_GPIO['Datetime'] = pd.to_datetime(df_Rust_GPIO['Datetime'])
df_Rust_GPIO["Language"] = "Rust GPIO"
# Rust Serial
df_Rust_Serial = pd.read_csv(file_name_Rust_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_Rust_Serial['Datetime'] = pd.to_datetime(df_Rust_Serial['Datetime'])
df_Rust_Serial['Result_ms'] = df_Rust_Serial['Result_ms'] - 0.9375
df_Rust_Serial["Language"] = "Rust Serial"
# C# Debug GPIO
df_CS_Deb_Remeas_GPIO = pd.read_csv(file_name_CS_Deb_Remeas_GPIO, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Deb_Remeas_GPIO['Datetime'] = pd.to_datetime(df_CS_Deb_Remeas_GPIO['Datetime'])
df_CS_Deb_Remeas_GPIO["Language"] = "C# Debug GPIO Remeasured"
# C# Debug Serial
df_CS_Deb_Remeas_Serial = pd.read_csv(file_name_CS_Deb_Remeas_Serial, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
df_CS_Deb_Remeas_Serial['Datetime'] = pd.to_datetime(df_CS_Deb_Remeas_Serial['Datetime'])
df_CS_Deb_Remeas_Serial['Result_ms'] = df_CS_Deb_Remeas_Serial['Result_ms'] - 0.9375
df_CS_Deb_Remeas_Serial["Language"] = "C# Debug Serial Remeasured"

# Tag all DataFrames
dfs = [
    df_CS_Deb_GPIO,
    df_CS_Rel_GPIO,
    df_CS_Deb_Remeas_GPIO,
    df_C_GPIO,
    df_Bash_GPIO,
    df_Python_GPIO,
    df_Rust_GPIO,
    df_CS_Deb_Serial,
    df_CS_Rel_Serial,
    df_CS_Deb_Remeas_Serial,
    df_C_Serial,
    df_Bash_Serial,
    df_Python_Serial,
    df_Rust_Serial,
]

# Concatenate dataframes
df_all = pd.concat(dfs, ignore_index=True)
df_all_gpio = pd.concat(dfs[:int(len(dfs)/2)], ignore_index=True)
df_all_serial = pd.concat(dfs[int(len(dfs)/2):], ignore_index=True)
```

```python
# === TIME SERIES PLOT ===
plt.figure(figsize=(18, 8))

sns.scatterplot(
    data=df_all,
    x="Datetime",
    y="Result_ms",
    hue="Language",
    palette="tab10",
    legend="full",
    edgecolor="none",
)

plt.title("Response time by Language and Interface (Compensated with message transmission time)", fontsize=16)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Response Time (ms)", fontsize=12)
plt.grid(True)
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(1))
plt.ylim(bottom=0)
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# === HISTOGRAM GPIO ===
plt.figure(figsize=(18, 8))

sns.histplot(
    data=df_all_gpio,
    x="Result_ms",
    hue="Language",
    bins=100,
    kde=True,
    element="step",
    stat="count",
    common_norm=False
)

plt.title("Combined Histogram of GPIO Response times by Language", fontsize=16)
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.xlim(left=0)
plt.xlim(right=8)
plt.tight_layout()
plt.show()

# === HISTOGRAM Serial ===
plt.figure(figsize=(18, 8))

sns.histplot(
    data=df_all_serial,
    x="Result_ms",
    hue="Language",
    bins=100,
    kde=True,
    element="step",
    stat="count",
    common_norm=False
)

plt.title("Combined Histogram of Serial port Response times by Language and Interface (Compensated with message transmission time)", fontsize=16)
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.xlim(left=0)
plt.xlim(right=20)
plt.tight_layout()
plt.show()
```

```python
# === HISTOGRAM GPIO (LOG) ===
plt.figure(figsize=(18, 8))

sns.histplot(
    data=df_all_gpio,
    x="Result_ms",
    hue="Language",
    bins=100,
    element="step",
    stat="count",
    common_norm=False
)

plt.title("Combined Histogram of GPIO Response times by Language on logarithmic scale", fontsize=16)
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.xlim(left=0)
plt.xlim(right=16)
plt.yscale("log")
plt.tight_layout()
plt.show()

# === HISTOGRAM Serial (LOG) ===
plt.figure(figsize=(18, 8))

sns.histplot(
    data=df_all_serial,
    x="Result_ms",
    hue="Language",
    bins=100,
    element="step",
    stat="count",
    common_norm=False
)

plt.title("Combined Histogram of Serial port Response times by Language on logarithmic scale (Compensated with message transmission time)", fontsize=16)
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.xlim(left=0)
plt.xlim(right=27)
plt.yscale("log")
plt.tight_layout()
plt.show()
```
