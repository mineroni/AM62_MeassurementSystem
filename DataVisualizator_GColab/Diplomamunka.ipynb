# C# Debug

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import statistics as stat

# Constants
file_name = "dataGpio_C#_Debug.csv"
meassurement = "GPIO pins"
language = "C# Debug mode"
fig_size = (18, 8)

# Load data from CSV file
df = pd.read_csv(file_name, sep=';', header=None, names=["ID", "UNIX Timestamp", "Datetime", "Result_ms"])
# Convert "Datetime" column to actual datetime object
df['Datetime'] = pd.to_datetime(df['Datetime'])

# === TIME SERIES PLOT ===
plt.figure(figsize=fig_size)
plt.scatter(
    df['Datetime'],
    df['Result_ms'],
    color="darkturquoise",
    s=1,
)
plt.title(f"Response time of {meassurement} in {language}")
plt.xlabel("Time")
plt.ylabel("Response time (ms)")
plt.grid(True)
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.show()

# === HISTOGRAM ===
plt.figure(figsize=fig_size)
sns.histplot(
    df['Result_ms'],
    bins=100,
    kde=True,
    color="darkturquoise",
    edgecolor=None,
).lines[0].set_color("darkcyan")
plt.title(f"Histogram of Response times meassuring {meassurement} in {language}")
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(0.2))
plt.show()

# === HISTOGRAM (LOG) ===
plt.figure(figsize=fig_size)
sns.histplot(df['Result_ms'], bins=100, color="darkturquoise")
plt.title(f"Histogram of Response times meassuring {meassurement} in {language} on logarithmic scale")
plt.xlabel("Response time (ms)")
plt.ylabel("Number of results")
plt.grid(True)
plt.yscale("log")
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.show()

# === BOXPLOT ===
plt.figure(figsize=fig_size)
sns.boxplot(y="Result_ms", data=df, color="darkturquoise")
plt.title(f"Boxplot of Response times meassuring {meassurement} in {language}")
plt.ylabel("Response time (ms))")
plt.grid(True)
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(0.5))
plt.show()


print(f"Average (átlag): {stat.mean(df['Result_ms'])}")
print(f"Median (medián): {stat.median(df['Result_ms'])}")
print(f"Mode (módusz): {stat.mode(df['Result_ms'])}")
print(f"Standard Deviation (szórás): {stat.stdev(df['Result_ms'])}")
print(f"Min (minimum): {min(df['Result_ms'])}")
print(f"Max (maximum): {max(df['Result_ms'])}")
print(f"Range (értékkészlet): {max(df['Result_ms']) - min(df['Result_ms'])}")
```
